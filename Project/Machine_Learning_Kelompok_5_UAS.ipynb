```python
from sqlalchemy import create_engine
from google.colab import userdata

password = userdata.get('Password')
user = userdata.get('User')
namaDb = "AnalisaKecelakaan"

simpan = f"postgresql://{user}:{password}@dataengineering5-dataengineering5.j.aivencloud.com:25498/{namaDb}?sslmode=require"

koneksi = create_engine(simpan).connect()

df_gabungan = pd.read_sql('SELECT * FROM kecelakaan ', koneksi)
```

```python
kolom_kategorik = ['Kabupaten_dan_Kota']
kolom_numerik = ['Curah_Hujan_Rendah', 'Curah_Hujan_Tinggi', 'Baik', 'Sedang', 'Rusak', 'Rusak_Berat', 'Jumlah_Kendaraan', 'Total_Kecelakaan']
```

```python
# # Menghitung matriks korelasi antar kolom numerik
korelasi = df_gabungan[kolom_numerik].corr()
korelasi
```

```python
# Visualisasi heatmap korelasi
fig = px.imshow(korelasi, text_auto=True, title='Korelasi antar Variabel')
fig.show()
```

```python
# Setup PyCaret untuk regresi
from pycaret.regression import *

reg = setup(
    data=df_gabungan.drop(columns=['Kabupaten_dan_Kota']),
    target='Total_Kecelakaan',
    session_id=123,
    normalize=True
)

# Bandingkan semua model → PyCaret pilih model terbaik
best_model = compare_models()
```

```python
# Evaluasi model terbaik
evaluate_model(best_model)
```

```python
# Tuning model terbaik
tuneup_model = tune_model(best_model)
```

```python
# Simpan model
save_model(tuneup_model, 'model_kecelakaan')
```

```python
# Prediksi hasil pada dataset
hasil = predict_model(tuneup_model)
hasil
```

```python
import plotly.express as px

# Visualisasi prediksi vs aktual
fig = px.scatter(hasil, x='Total_Kecelakaan', y='prediction_label', title='Aktual vs Prediksi Jumlah Kecelakaan', trendline='ols')
fig.show()
```
